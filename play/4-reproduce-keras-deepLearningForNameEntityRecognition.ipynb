```python
# imports
import pandas as pd

from tqdm import tqdm
```

```python
# Load and Update Data
def load_data(file_path, num_samples):
    sentence_idxs = []
    words = []
    ner_tags = []
    
    with open(file_path, 'r') as f:
        lines = f.read().split("\n") # separate each sentence by new line
        # print(lines)
    sentence_idx = 0
    for line in lines:
        # text_idx, input_text, text_ner_tag = line.split("[")
        
        each_line = line.split(" ")
        # print(len(each_line))
        if len(each_line) != 1:
            sentence_idx += 1
            text_idx, input_text, ner_tag = each_line[0], each_line[1], each_line[2]
            print(f"Line {text_idx} has word {input_text}, with NER Tag {ner_tag} --- {sentence_idx}")
            sentence_idxs.append(sentence_idx)
            words.append(input_text)
            ner_tags.append(ner_tag)
        else:
            # print(f"Line {text_idx} is empty")
            sentence_idx = 0
            reset_sentence_idx, input_text, ner_tag = " ", " ", " "
            sentence_idxs.append(reset_sentence_idx)
            words.append(input_text)
            ner_tags.append(ner_tag)
            print()

    return sentence_idxs, words, ner_tags
```

```python
train_file_path = "data/train"
N_samples = 10000
sentence_idxs, train_words, train_ner_tags = load_data(train_file_path, N_samples)
```

```python
import tqdm
def sentence_integrate(data):
  agg_func = lambda s: [(w, p, t) for w, p, t in zip(s["Index"].values.tolist(),
                                                           s["Word"].values.tolist(),
                                                           s["NER Tag"].values.tolist())]
  return data.groupby('Index').apply(agg_func).tolist()
  
sentences=sentence_integrate(word_per_tag_df)
```

```python
def create_mappings(words, per_tags):  
    word2idx = {}
    tag2idx = {}

    # Constructing word2idx dictionary
    for i, w in enumerate(words):
        word2idx[w] = i + 1

    # Constructing tag2idx dictionary
    for i, t in enumerate(per_tags):
        tag2idx[t] = i

    # Print the dictionaries
    # print("word2idx:", word2idx)
    # print("tag2idx:", tag2idx)

    return word2idx, tag2idx
```

```python
word_mappings, ner_tag_mappins = create_mappings(train_words, train_ner_tags)
```

```python
words = list(set(word_per_tag_df["Word"].values))
words.append("ENDPAD")
num_words = len(words)
num_words
```

```python
# num_words = len(word_per_tag_df['Word'].values)
# print(num_words)

# num_words = len(word_mappings)
# print(num_words)
max_len = 50

# Print out the sentences to check for any anomalies
# print(sentences)

X = [[word_mappings[w[1]] for w in s] for s in sentences]
X = pad_sequences(maxlen=max_len, sequences=X, padding="post", value=num_words-1)

y = [[ner_tag_mappins[w[2]] for w in s] for s in sentences]
y = pad_sequences(maxlen=max_len, sequences=y, padding="post", value=ner_tag_mappins["O"])

```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

#build model
from tensorflow.keras import Model, Input
from tensorflow.keras.layers import LSTM, Embedding, Dense
from tensorflow.keras.layers import InputLayer, TimeDistributed, SpatialDropout1D, Bidirectional
from tensorflow import keras

model = keras.Sequential()
model.add(InputLayer((max_len)))
model.add(Embedding(input_dim=num_words, output_dim=max_len, input_length=max_len))
model.add(SpatialDropout1D(0.1))
model.add( Bidirectional(LSTM(units=100, return_sequences=True, recurrent_dropout=0.3)))

#model summary
model.summary()
```

```python
len(x_train), len(x_test), len(y_train), len(y_test)
```

```python
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from livelossplot.tf_keras import PlotLossesCallback

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.random.seed(0)
# plt.style.use("ggplot")

import tensorflow as tf
# print('Tensorflow version:', tf.__version__)
# print('GPU detected:', tf.config.list_physical_devices('GPU'))

# %%time

# logdir="log/"
# tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)

chkpt = ModelCheckpoint("model_weights.h5", monitor='val_loss',verbose=1, save_best_only=True, save_weights_only=True, mode='min')

early_stopping = EarlyStopping(monitor='val_accuracy', min_delta=0, patience=1, verbose=0, mode='max', baseline=None, restore_best_weights=False)

callbacks = [PlotLossesCallback(), chkpt, early_stopping]


history = model.fit(
    x=x_train,
    y=y_train,
    validation_data=(x_test,y_test),
    batch_size=32, 
    epochs=3,
    callbacks=callbacks,
    verbose=1
    
)
```
